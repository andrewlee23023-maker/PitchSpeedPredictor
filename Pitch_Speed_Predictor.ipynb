```python
import pandas as pd
import io
import numpy as np
df = pd.read_csv("https://raw.githubusercontent.com/drivelineresearch/openbiomechanics/main/baseball_pitching/data/poi/poi_metrics.csv")
print(df)
```

```python
df.drop(df.columns[:4],axis= 1, inplace = True)
print(df)
```

```python
df.dropna(how = 'any',inplace=True)
y_np = df.iloc[:, 0].to_numpy()
df.drop(df.columns[0], axis=1, inplace=True)
print(y_np)
```

```python
x_np = df.to_records()
print(x_np)
print(x_np.dtype.names)
```

```python
import torch
import torch.nn as nn
from torch.utils.data import random_split, DataLoader, TensorDataset
import numpy as np
from sklearn.preprocessing import StandardScaler

net = nn.Sequential(
    nn.Linear(76,256),
    nn.BatchNorm1d(256),
    nn.ELU(),
    nn.Dropout(p=float(0.1)),
    nn.Linear(256,128),
    nn.BatchNorm1d(128),
    nn.ELU(),
    nn.Dropout(p=float(0.2)),
    nn.Linear(128,1)
)

# Standardize input features
scaler = StandardScaler()
x_scaled = scaler.fit_transform(np.array(x_np.tolist())[:, 1:])
x = torch.tensor(x_scaled, dtype = torch.float32)
y = torch.tensor(y_np, dtype = torch.float32)

total = 403
train = int(total*0.8)
test = total-train
dataset = TensorDataset(x, y)
train_dataset, test_dataset = random_split(dataset, [train, test])
batch_size = 100
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)
```

```python
import torch.optim as optim
loss = nn.MSELoss()
optimizer = optim.SGD(net.parameters(),lr = 0.0001)
```

```python
epochs = 2000
for epoch in range(epochs):
  running_loss = 0
  for i, data in enumerate(train_loader, 0):
    inputs, labels = data
    optimizer.zero_grad()
    outputs = net(inputs)
    calculated_loss = loss(outputs.squeeze(),labels)
    calculated_loss.backward()
    optimizer.step()
    running_loss += calculated_loss.item()
  if epoch % 100 == 99:
    print(f'Epoch {epoch + 1}, Loss: {running_loss / len(train_loader):.3f}')

print('Finished Training')

net.eval()
test_loss = 0.0
with torch.no_grad():
  for data in test_loader:
    inputs, labels = data
    outputs = net(inputs)
    current_test_loss = loss(outputs, labels.unsqueeze(1))
    test_loss += current_test_loss.item()

print(f'Test Loss: {test_loss / len(test_loader):.3f}')
```

```python
import torch.nn.functional as F

net.eval() # Set the model to evaluation mode
total_abs_error = 0.0
num_samples = 0

with torch.no_grad():
    for inputs, labels in test_loader:
        outputs = net(inputs)
        # Calculate L1 Loss (MAE)
        abs_error = F.l1_loss(outputs, labels.unsqueeze(1), reduction='sum')
        total_abs_error += abs_error.item()
        num_samples += labels.size(0)

mean_absolute_error = total_abs_error / num_samples
print(f'Average Absolute Error (MAE): {mean_absolute_error:.3f} mph')
```
